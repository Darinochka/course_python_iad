### Задание 10. Ответьте на вопросы:

* Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике? 

```python
# Какая доля студентов сдала экзамен по математике?
len(df[df['math score'] >= passmark]) / len(df)
```

```python
# Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?'
condition = df['test preparation']  == 'completed'
len(df[condition & (df['math score'] >= passmark)]) / len(df[condition])
```

```python
# Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?
condition = (df['gender'] == 'female') & (df['test preparation'] == 'none')
len(df[condition & (df['math score'] < passmark)]) / len(df[condition])
```

### Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.

```python
%%time
# для каждой этнической группы выведите средний балл за экзамен по чтению

mean_race = {race : [0, 0] for race in df['race/ethnicity'].unique()}
for i in range(len(df)):
  mean_race[df.loc[i]['race/ethnicity']][0] += 1
  mean_race[df.loc[i]['race/ethnicity']][1] += df.loc[i]['reading score']

print(dict(map(lambda x: (x[0], x[1][1] / x[1][0]), mean_race.items())))
```

С использованием groupby - 6.9ms

С использованием циклов - 480ms

В почти 70 раз быстрее!

```python
%%time
# для каждого уровня образования выведите минимальный балл за экзамен по письму

min_educ = {educ: 101 for educ in df['education'].unique()}
for i in range(len(df)):
  educ_st = df.loc[i]['education']
  score_st = df.loc[i]['writing score']
  
  if min_educ[educ_st] > score_st:
    min_educ[educ_st] = score_st

print(min_educ)
```

### Задание 16. Переведем баллы в оценки

**Система перевода баллов в оценки**

больше 90 = A

80-90 = B

70-80 = C

60-70 = D

50-60 = E

меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

Проверю максимальную оценку для каждого предмета

```python
df.max()
```

```python
def convert_grade(score):
  for sc, grade in zip(range(50, 110, 10), 'FEDCBA'):
    if score <= sc:
      return grade
```

Получу серию средних оценок для каждого студента

```python
def mean_score(student):
  return student.mean()

mean_scores = df[['math score', 'reading score', 'writing score']].apply(mean_score, axis=1)
```

Теперь применю на эту серию функцию

```python
df['Grade'] = mean_scores.apply(convert_grade)
```
